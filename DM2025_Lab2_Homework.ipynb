# **Competition Code**

## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section

import pandas as pd
import re
import json

# 讀取資料
with open("/content/drive/MyDrive/Colab Notebooks/Phace3 data/final_posts.json", "r", encoding="utf-8") as f:
    raw = json.load(f)

posts = []
for item in raw:
    p = item["root"]["_source"]["post"]
    posts.append({
        "id": p["post_id"],
        "text": p["text"],
        "hashtags": " ".join(p.get("hashtags", []))
    })

df_posts = pd.DataFrame(posts)
df_emotion = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Phace3 data/emotion.csv")
df_split   = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Phace3 data/data_identification.csv")

df = df_posts.merge(df_split, on="id").merge(df_emotion, on="id", how="left")
train_df = df[df.split == "train"].copy()
test_df  = df[df.split == "test"].copy()

# 文字清理（保留emojiy字串&hashtag）
def clean(t, h=""):
    if pd.isna(t): t = ""
    if pd.isna(h): h = ""
    s = f"{t} {h}"
    s = re.sub(r"http\S+|www\.\S+|@\w+", " ", s) # 移除網址與@
    return re.sub(r"\s+", " ", s).strip()

train_df["text"] = train_df.apply(lambda x: clean(x.text, x.hashtags), axis=1)
test_df["text"]  = test_df.apply(lambda x: clean(x.text, x.hashtags), axis=1)
```

## 2. Feature Engineering Steps

```python
### Add the code related to the feature engineering steps in cells inside this section

from sklearn.feature_extraction.text import TfidfVectorizer

#TF-IDF 特徵（ngram=1~3，sublinear_tf能讓高頻詞影響變小）
vectorizer = TfidfVectorizer(
    ngram_range=(1, 3),
    max_features=100000,
    sublinear_tf=True,
    lowercase=True,
    stop_words=None          # 不移除中文停用詞（效果反而比較好）
)

X_train = vectorizer.fit_transform(train_df.text)
X_test  = vectorizer.transform(test_df.text)
y_train = train_df.emotion

print(f"TF-IDF: {len(vectorizer.vocabulary_)}")
print(f"特徵矩陣形狀: {X_train.shape}")
```

## 3. Model Implementation Steps

```python
### Add the code related to the model implementation steps in cells inside this section
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, f1_score
import numpy as np

# 關鍵：class_weight='balanced' 自動處理類別不平衡
model = LinearSVC(
    C=1.0,
    class_weight='balanced',   # 讓 disgust 這種少數類不會被忽略
    max_iter=1000,
    random_state=42
)

model.fit(X_train, y_train)

# 產生 submission
predictions = model.predict(X_test)

submission = pd.DataFrame({
    "id": test_df.id,
    "emotion": predictions
})

submission.to_csv("submission_tfidf_linearsvc_balanced.csv", index=False)

# 預測分佈（可檢查是否太偏 joy）
print(submission.emotion.value_counts())
```
